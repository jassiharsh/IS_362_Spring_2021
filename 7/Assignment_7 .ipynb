# Movie Ratings


### 1. Load the ratings by user information that you collected into a pandas dataframe.

```python
import pandas as pd

# import csv to pandas
data=pd.read_csv("movies.csv", delimiter=",") 
data=data.set_index('user')
data
```

### 2. Show the average ratings for each user and each movie

- Step 1 Reshape data

```python
df=pd.melt(data.reset_index(), id_vars=['user'], value_vars=list(data.columns)[0:], var_name='movie', value_name='rating')
df.head(10)
```

- step 2 Calucuate average ratings

```python
#average user ratings
df1 = df.groupby(['user'])['rating'].agg(['count','mean'])
df1
```

```python
#average movie rating
df1 = df.groupby('movie')['rating'].agg(['count','mean'])
df1
```

### 3. Create a new pandas dataframe, with normalized ratings for each user. Again, show the average ratings for each user and each movie.

- Create a normalized dataframe

```python
normalized_df=(data-data.min())/(data.max()-data.min())
normalized_df
```

#### - average user ratings

```python
#reshape data
normalized_df=pd.melt(normalized_df.reset_index(), id_vars=['user'], value_vars=list(data.columns)[0:], var_name='movie', value_name='rating')

#average user ratings
df3 = normalized_df.groupby('user')['rating'].agg(['count','mean'])
df3
```

#### - average movie ratings

```python
#average movie ratings
df4 = normalized_df.groupby('movie')['rating'].agg(['count','mean'])
df4
```

### 4. Advantages of using Normalized ratings instead of actual ratings

####  1. Advantages
    - Boundaries can be set and all indicators have an identical range (0, 1)
#### 2. Disadvantages
    - Values do not maintain proportionability
    - Difference in variance is not fully eliminated

### 5. Create another new pandas dataframe, with standardized ratings for each user. Once again, show the average ratings for each user and each movie.

```python
#standardized dataframe
standard_df = (data-data.mean())/data.std()
standard_df
```

- average user ratings

```python
#reshape data
standard_df=pd.melt(standard_df.reset_index(), id_vars=['user'], value_vars=list(data.columns)[0:], var_name='movie', value_name='rating')

#average user ratings
df5 = standard_df.groupby('user')['rating'].agg(['count','mean'])
df5
```
